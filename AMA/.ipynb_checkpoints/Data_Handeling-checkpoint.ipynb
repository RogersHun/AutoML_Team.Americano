# Ford A Data Set

```python
def read_ariff(path):
    raw_data, meta = loadarff(path)
    cols = [x for x in meta]
    data2d = np.zeros([raw_data.shape[0],len(cols)])
    
    for i,col in zip(range(len(cols)),cols):
        data2d[:,i] = raw_data[col]
    return data2d

train = read_ariff("./FordA_TRAIN.arff")
test = read_ariff("./FordA_TEST.arff")
```

# Pre-Process Ph, Temp Data Set

```python
date = pd.date_range('2021-09-06','2021-10-27',freq='1D')
```

```python
date 
```

```python
date_result = []
arr = []
for i in date:
    d=str(i)
    d=d.replace(' 00:00:00','')
    arr.append(d)
    d = d.replace('-','.')
    d = "kemp-abh-sensor-"+d+".csv"
    date_result.append(d)
date_result
```

```python
for i in range(len(date_result)):
    try:
        if date_result[i] =="kemp-abh-sensor-2021.09.06.csv":
            init_data = pd.read_csv("./dataset/%s"%date_result[i])
            init_data['Date'] = arr[i]
        else:
            data = pd.read_csv("./dataset/%s"%date_result[i])
            data['Date'] = arr[i]
            init_data = pd.concat([init_data,data],ignore_index=True)
            

    except:
        continue
    
init_data['Process Rate']=0
```

```python
init_data
```

```python
pr = pd.read_csv('./dataset/pr.csv')
```

```python
db = pd.merge(init_data, pr, on=['Date', 'LoT'])
```

```python
del db['Index'], db['Process Rate_x']
```

```python
db = db.rename(columns={'Process Rate_y':'PR'})
```

```python
db
```

```python
db.to_csv("./Tem_pH_Data.csv",index=False, encoding="utf-8-sig")
```
